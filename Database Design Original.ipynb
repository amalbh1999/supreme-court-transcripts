```python
justices = re.findall('JUSTICE[A-Z\s]+:',text)
```

```python
def unique(lst):
    uni = []
    for i in lst:
        if i not in uni:
            uni += [i]
    return uni
```

```python
def case_no(txt):
    return unique(re.findall('No.\s\d+-\d+',text))
```

```python
case_no(text1)
```

```python
def justices(texts):
    d = {}
    for txt in texts:
        j = re.findall('JUSTICE[A-Z\s]+:',txt)
        justice = unique(j)
        num = case_no(txt)[0]
    for n in num:
        d[n] = justice
    return justice
```
